# Python Project: Evaluating Stock Prices
#### We will be evaluating th estock prices for Apple (APL) and AMG

```python
import yfinance as yf
import pandas as pd
```

```python
apple = yf.Ticker("AAPL")
```

```python
# This gives us the infor surrounding apple in a dictionary format
apple_info=apple.info
apple_info
```

```python
#We call on the country key to see what country apple is located in (HQ)
apple_info['country']
```

```python
# We will be extracting the share price of apple over their history
# We set period over max so that we get the data over the max period of time
apple_share_price_data = apple.history(period="max")
```

```python
# Calls only the first 5 rows of the dataframe
apple_share_price_data.head()
```

```python
# Using the data.reset_index function we can reset the index of the data frame
apple_share_price_data.reset_index(inplace=True)
```

```python
apple_share_price_data.plot(x="Date", y="Open")
```

```python
# Using the 'dividends' variable we are able to extract the dividends
apple.dividends
```

```python
#This. graph plots the dividends over a period of time
apple.dividends.plot()
```
